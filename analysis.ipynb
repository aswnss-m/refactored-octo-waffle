```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./runner_result.csv")
df = pd.DataFrame(df).dropna(axis=0)
```

```python
plt.figure(figsize=(6, 3))  # Adjust the width and height as needed
plt.xlabel("steps")
plt.ylabel("headway")
plt.plot(df.headway)
plt.show()
mean = df.headway.mean()
print("Mean headway : ",mean)
```
